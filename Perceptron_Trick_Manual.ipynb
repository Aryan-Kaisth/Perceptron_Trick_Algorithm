```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state = 3, hypercube=False,class_sep=10)
```

```python
X.shape
```

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.colors import ListedColormap
plt.figure(figsize=(10,6))

cmap = ListedColormap(["red", "green"])

plt.scatter(X[:,0], X[:,1], c=y, cmap=cmap, s=100)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter plot")
plt.show()
```

```python
def stepfunc(z):
    if z >= 0:
        return 1
    else:
        return 0
```

```python
def perceptron(X,y):

    X = np.insert(X,0,1,axis=1) # for bias term
    weights = np.array([1, 1, 1]) # initialize weights
    learningrate = 0.1 # learning rate

    # main algorithm
    for _ in range(1000): # epoch
        i = np.random.randint(0,X.shape[0]) # Selecting any row from X
        y_pred = stepfunc(np.dot(X[i],weights)) # Giving 0 ans 1
        weights = weights + learningrate*(y[i]-y_pred)*X[i]

    return weights[0],weights[1:] # to get intercept and coefficients
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(intercept_)
print(coef_)
```

```python
# from coef_ we get two points a and b
# from intercept we get c
m = -(coef_[0]/coef_[1]) # m = -a/b
b = -(intercept_/coef_[1]) # c = -c/b
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b # y = mx+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap=cmap,s=100)
plt.show()
```

```python
from sklearn.linear_model import Perceptron
```

```python
model = Perceptron()
```

```python
model.fit(X,y)
```

```python
coef_perceptron = model.coef_[0]
intercept_perceptron = model.intercept_
```

```python
coef_perceptron
```

```python
m_perceptron = -(coef_perceptron[0]/coef_perceptron[1]) # m = -a/b
b_perceptron = -(intercept_perceptron/coef_perceptron[1]) # c = -c/b
```

```python
x_input_perceptron = np.linspace(-3,3,100)
y_input_perceptron = m_perceptron*x_input_perceptron + b_perceptron # y = mx+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input_perceptron,y_input_perceptron,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap=cmap,s=100)
plt.show()
```
